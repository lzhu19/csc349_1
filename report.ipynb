```python
import pandas as pd
import matplotlib.pyplot as plt
```

Read in the data from 'results.csv':

```python
df = pd.read_csv('results.csv')
df.head()
```

### **Scatterplot**

```python
plt.scatter(df['Size'], df['Selection'], label='Selection Sort')
plt.scatter(df['Size'], df['Merge'], label='Merge Sort')
plt.scatter(df['Size'], df['Counting'], label='Counting Sort')
plt.title('Algorithm Runtime')
plt.xlabel('Sequence Size')
plt.ylabel('Running Time (ms)')
plt.legend(loc='upper left', title='Legend')
plt.show()
```

### **Comparison**
Check for 300 data points total.

All three sorting algorithms had 100 non-null values, meaning there are 3*100= 300 data points total.

### **Brief Analysis**

The curve for **selection sort** matches $O(n^2)$ because this algorithm iterates through the unsorted list to find the minimum value. In the worst case, the minimum value is at the end of the unsorted list, which makes this algorithm really bad for larger sequence sizes.

The curve for **merge sort** matches $O(nlogn)$ because the points on the graph follow a $y = nlogn$ curve. Merge sort does better than selection sort because it uses divide & conquer to solve smaller tasks in a more efficient way.

The curve for **count sort** matches $O(n+k)$ because it accesses values directly using indices, making it very effective for larger sequence sizes. This algorithm does relatively worse in smaller sequence sizes because it still needs to iterate through the sequence to generate a list of frequencies.
